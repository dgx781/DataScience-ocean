```python
from sklearn.impute import SimpleImputer
imputer= SimpleImputer(missing_values= np.nan, strategy="median")
df[num_col]= imputer.fit_transform(df[num_col])
```

```python
l_lim= df['Sale Price'].quantile(0.25)
l_lim
```

```python
u_lim= df['Sale Price'].quantile(0.75)
u_lim
```

```python
def limit_imputer(val):
    if val<l_lim:
        return l_lim
    if val>u_lim:
        return u_lim
    else:
        return val
```

```python
df['Sale Price']= df['Sale Price'].apply(limit_imputer)
df['Sale Price'].head()
```

```python
mp= {'Yes':1,
    'No':0}
```

```python
from sklearn.preprocessing import StandardScaler
scaler= StandardScaler()
x_train= scaler.fit_transform(x, y)
```

```python
from sklearn.linear_model import LinearRegression
lr= LinearRegression(normalize= True)
lr.fit(x_train, y_train)
```

```python
x_test['Waterfront View']= x_test['Waterfront View'].map(mp)
x_test['Ever Renovated']= x_test['Ever Renovated'].map(mp) 
x_test.head()
```

```python
residuals= pred- y_test
residual_table= pd.DataFrame({'residual': residuals, 'predictions': pred})
residual_table= residual_table.sort_values(by= 'predictions')
residual_table.head()
```

```python
z= [i for i in range(int(residual_table['predictions'].max()))]
k= [0 for i in range(int(residual_table['predictions'].max()))]
```

```python
plt.figure(dpi= 120, figsize= (17,7))
plt.scatter(residual_table['predictions'], residual_table['residual'], color= "skyblue", s=2)
plt.plot(z,k, color="red", label= "Regression Line")
plt.ylim(-800000,800000)
plt.xlabel("Fitted points(ordered by predictions)")
plt.ylabel("residuals")
plt.title("Residual Plot")
plt.legend()
plt.show
```
