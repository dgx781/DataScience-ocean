# Meshgrid

```python
x= np.random.randint(0,50,7)
y= np.random.randint(-10,10,5)
xval, yval= np.meshgrid(x,y)
print("X values= ")
print(xval)
print("Y values= ")
print(yval)
```

# Quiver Plot

```python
x= np.linspace(3,7,10)
y= np.linspace(-3,15,8)
xval, yval= np.meshgrid(x,y)
v= np.tan(xval)
u= np.tanh(yval)
plt.quiver(xval,yval,u,v)
```

# Contour Plot

```python
x= [-6,-4,-2,0,2,4,6]
y= [-3,-2,-1,0,1,2,3]
xval, yval= np.meshgrid(x,y)
zval= np.cos(xval)*np.sin(yval)
plt.contour(xval,yval,zval)
```
